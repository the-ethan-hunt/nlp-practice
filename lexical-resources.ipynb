This is a simple use of the lexical resources in nltk.

```python
import nltk
```

```python
from nltk.corpus import udhr
```

UDHR or the Universal Declaration of human rights is a corpus by the United Nations with 480k words in 300 languages

```python
udhr.fileids()
```

```python
languages=['Afrikaans','Catalan','Frisian','Spanish','Kinyarwanda','Quechua','Shona','Swaheli','Zulu']
```

We have selected some random languages to plot a distribution curve to show the number of letters in each word in the given language

```python
cfd = nltk.ConditionalFreqDist(
(lang,len(word))
for lang in languages
for word in udhr.words(lang+'-Latin1'))
```

```python
cfd.plot(cumulative=True)
```

It is observed that most of these languages have a majority count at 5 or 6 letter words with the exceptions of Quechua, Shona and Zulu
